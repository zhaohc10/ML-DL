# Run SVC model

## load dataset

```python
%matplotlib inline 
#%matplotlib qt
import pandas as pd
path=r'noExpired.xlsx'
all=pd.read_excel(path)
```

## Manipulate Numeric Feature

```python
from sklearn.preprocessing import MinMaxScaler
from __future__ import division

scaler = MinMaxScaler()

numeric_cols = ['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'number_outpatient', 'number_emergency', 'number_inpatient', 'number_diagnoses']
all[ numeric_cols ] = all[ numeric_cols ].astype(float)
all[ numeric_cols ]=all[ numeric_cols ].apply(lambda x: MinMaxScaler().fit_transform(x))
all.head(10)
```

## Manipulate Categorical Feature

```python
x_num_all = all[ numeric_cols ].as_matrix()
# categorical
cat_all = all.drop( numeric_cols + [ 'readmitted'], axis = 1 )

fac_x_cat_all = pd.DataFrame()
cat_cols = list(cat_all.columns.values)
for col in cat_cols:
    all_cur, _ = pd.factorize(cat_all[col])
    fac_x_cat_all[col] = all_cur

fac_x_cat_all = fac_x_cat_all.as_matrix()

import numpy as np
x_all = np.hstack(( x_num_all, fac_x_cat_all ))
```

```python
from sklearn.preprocessing import label_binarize
# y
y_all = all.readmitted
```

## Split dataset to Train and Test

```python
from sklearn.cross_validation import train_test_split
#train-test split
x_train, x_test, y_train, y_test = train_test_split(x_all, y_all, test_size=.25,random_state=24)
```

## Train SVM model when parameter is default

```python
import time
from sklearn.svm import SVC
# although one needs to choose these hyperparams
#training
st = time.time()
print "training started"

C = 173
gamma = 1.31e-5
shrinking = True
probability = True
verbose = True
#kernel="linear"
svc = SVC( C = C, gamma = gamma, shrinking = shrinking, probability = probability, verbose = verbose )
svc.fit( x_train, y_train )

"""
from sklearn.metrics import euclidean_distances
svc = SVC(kernel="linear")
svc.fit( x_train, y_train )
w = svc.dual_coef_.dot(svc.support_vectors_)
margin = 2 / np.sqrt((w ** 2).sum())
distances = euclidean_distances(neg_samples, pos_samples)
assert np.abs(margin - distances.min()) < 1e-6
"""


print "training ended"
et = time.time()
tt = et - st
print "Training Time = " + str(tt) + "\n"
```

### Parameters of  trained SVC model 
#### We can tentatively conclude that svc.dual_coef_ corresponds to alpha * y 

```python
w = svc.dual_coef_ 
b = svc.intercept_[0] 
ze=0.00000000e+00
numZeros=sum(1 for i in w.flat if i==ze)
print "Number of zeros: %d" %numZeros
print "The weight of dual problem and the bias:"
w,b
```

```python
# get support vectors
print 'Support vectors '
svc.support_vectors_
```

```python

# get indices of support vectors
print 'Indices of support vectors:'
svc.support_ 
```

```python

# get number of support vectors for each class
print 'Number of support vectors for each class'
svc.n_support_ 
```

### Score of trained SVC model

```python
print 'Score of trained LR model %.3f' %svc.score(x_train,y_train)
```

### ROC and AUC of trained SVC model

```python
from sklearn.metrics import precision_recall_curve, roc_curve,auc
#predictions
pred = svc.predict( x_test )

fpr, tpr, thresholds =roc_curve(y_test, pred)
roc_auc = auc(fpr, tpr)
print("Area under the ROC curve : %f" % roc_auc)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc) 
plt.plot([0, 1], [0, 1], 'k--') 
plt.xlabel('False Positive Rate') 
plt.ylabel('True Positive Rate') 
plt.legend(bbox_to_anchor=(0.9, 0.2), 
    bbox_transform=plt.gcf().transFigure) 
plt.show()
```

## Find best SVC Model using conventional GridSearchCV

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import StratifiedKFold
from sklearn.grid_search import GridSearchCV

param_grid = {'penalty':['l1','l2'],'C': [0.001, 0.01, 0.1, 1,10,100,1000,1e4,1e5,1e6] }
lr=LogisticRegression()
clf = GridSearchCV(lr, 
                   param_grid,
                   scoring="accuracy",
                   cv=StratifiedKFold(y_all, n_folds=10),
                   verbose=1)

%time clf.fit(x_all, y_all)
```

### The best score and corresponding parameters after using GridSearchCV

```python
clf.best_score_, clf.best_params_
```

### The weight, bias and number of zeros of best LR moedel after using GridSearchCV

```python
w_best=clf.best_estimator_.coef_ 
b_best=clf.best_estimator_.intercept_[0] 
ze=0.00000000e+00
numZeros=sum(1 for i in clf.best_estimator_.coef_.flat if i==ze)
print "Number of zeros: %d" %numZeros
w_best,b_best
```

### ROC using suitable LR model after using GridSearchCV

```python
pred_opt = clf.predict( x_test )
fpr, tpr, thresholds =roc_curve(y_test, pred_opt)

roc_auc = auc(fpr, tpr)
print("Area under the ROC curve : %f" % roc_auc)

plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc) 
plt.plot([0, 1], [0, 1], 'k--') 
plt.xlabel('False Positive Rate') 
plt.ylabel('True Positive Rate') 
plt.legend(bbox_to_anchor=(0.9, 0.2), 
    bbox_transform=plt.gcf().transFigure) 
plt.show()
```
